# **ANALYSIS 01 : Descriptive Statistics:**

Calculation of basic descriptive statistics :
*   Mean
*   Median
*   Mode
*   Standard Deviation
*   Variance

This will help us to understand the central tendency and variability in the data.

```python
import pandas as pd

data = {
    'Hours': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8],
    'Scores': [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Calculate descriptive statistics
scores_stats = df['Scores'].describe()

# Calculate mode separately since it's not included in the describe() output
scores_mode = df['Scores'].mode().values[0]

# Calculate variance separately since it's not included in the describe() output
scores_variance = df['Scores'].var()

print("MARKS STATISTICS SUMMARY IS AS BELOW:")
print(scores_stats)
print("Scores Mode:", scores_mode)
print("Scores Variance:", scores_variance)
```

# **ANALYSIS 03 : Pearson correlation coefficient**
**Correlation coefficients are typically interpreted as follows:**


*  Strong correlation: A correlation coefficient of 0.7 or higher ==> Strong correlation
*   Moderate correlation: A correlation coefficient of 0.3 to 0.7 ==> Moderate correlation
*   Moderate correlation: A correlation coefficient of 0.3 to 0.7 ==> Moderate correlation
*   Weak correlation: A correlation coefficient of 0 to 0.3 ==> Weak correlation


```python
# Calculate the Pearson correlation coefficient
correlation_coefficient = df['Hours'].corr(df['Scores'])

print(f"Pearson Correlation Coefficient: {correlation_coefficient:.2f}")
```

# **ANALYSIS 04 : Regression Analysis**

**In our Regression Model is followed by these step:**


1.   Dependent variable (y) = 'Scores (out of 20)'
2.   Independent variable (X) = 'Study Time (Minutes)'.
3.   We add a constant term () to the independent variable to account for the intercept in the linear regression model.
4.   We fit a linear regression model using sm.OLS() from statsmodels.

```python
import statsmodels.api as sm

# Define the dependent variable (y) and the independent variable (X)
y = df['Scores']
X = df['Hours']

# Add a constant to the independent variable (X) for the intercept term
X = sm.add_constant(X)

# Fit a linear regression model
model = sm.OLS(y, X).fit()

# Get the model summary
print(model.summary())
```

```python
# Plotting the distribution of scores
import matplotlib.pyplot as plt

df.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```
