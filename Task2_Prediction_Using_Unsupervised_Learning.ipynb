From the given 'Iris' dataset, predict the optimum number of clusters and represent it visually.

```python
# Importing all the necessary required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

#Inserting the Inb-byilt IRIS dataset

iris = datasets.load_iris()
iris  #To view complete dataset iris
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names) # conevrting it to tabular form
iris_df
```

```python
iris_df.head()   # To view only the first 5 rows
```

```python
# Finding the optimum number of clusters for k-means classification

x = iris_df.iloc[:, [0, 1, 2, 3]].values
x
```

```python
# Viewing X in Tabular form

x2= pd.DataFrame(iris_df.iloc[:, [0, 1, 2, 3]].values)
x2
```

```python
from sklearn.cluster import KMeans

wcss = [] # This list will store the "Within-Cluster Sum of Squares" (WCSS) values for different numbers of clusters.

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

```

*  ** n_clusters=i:**
This parameter sets the number of clusters the K-means algorithm should create. It's set to the current value of i in the loop.

*   **init='k-means++':**

 This specifies the method for initializing the initial cluster centers. 'k-means++' is a smart initialization technique that tends to give better results.

*   **max_iter=300: **

This parameter determines the maximum number of iterations the K-means algorithm will perform for a single run.

*   **n_init=10: **

The number of times the algorithm will be run with different initial seeds. This helps in finding the best solution among multiple runs.

*   **random_state=0:**

This is a random seed to ensure reproducibility. Setting it to a specific value (0 in this case) makes the results consistent across different runs.

```python
# Plotting the results onto a line graph,
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow Graph')
plt.xlabel('Number of clusters = k ')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
print(y_kmeans.shape)
print(y_kmeans[:10])  # Print the first 10 elements of y_kmeans
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
